```python
# Primero cargamos los csv necesarios, empecemos con el set de datos de test
dfTest = pd.read_csv('/home/matias/Escritorio/ML/data.csv')
dfTest.info()
```

```python
dfTest= dfTest[['place_name','price']]
dfTest = dfTest[:5000]
precioPorBarrio = dfTest.groupby('place_name').agg([np.mean,np.std])
precioPorBarrio = precioPorBarrio[10:30]
precioPorBarrio.head()
```

```python
%matplotlib inline
precioPorBarrio.boxplot(column='price', by='place_name', figsize=(20,32))
```
