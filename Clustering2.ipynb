```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import median_absolute_error
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import Imputer
```

```python
properties = pd.read_csv('/home/santiago/Desktop/Organizacion De Datos (75.06)/TP2/trainPreProcess.csv', low_memory=False)
test_set = pd.read_csv('/home/santiago/Desktop/Organizacion De Datos (75.06)/TP2/testPreProcess.csv', low_memory=False)
```

```python
#Me quedo con las columnas que me interesan para entrenar el algoritmo
train = properties[['id','created_on','property_type_encoded', 'state_name_encoded', 'place_name_encoded', 'rooms', 
                    'surface_covered_in_m2', 'floor','distancia_obelisco','distancia_minima_subte',
                    'distancia_minima_privado','distancia_minima_publico','price']]
```

```python
#Tomo una muestra con el 80% de los elementos del set de entrenamiento
train = train[0:int(0.8*len(train))]
train_sub_test = train[int(0.8*len(train)):]
```

```python
#Remuevo al set de entrenamiento los precios de las propiedades
train_no_prices = train.loc[:, ['property_type_encoded', 'state_name_encoded', 'place_name_encoded', 'rooms', 'floor'
                                ,'distancia_obelisco','distancia_minima_subte','distancia_minima_privado',
                                'distancia_minima_publico']]
```

```python
sub_train_to_predict = train_sub_test.loc[:, ['property_type_encoded', 'state_name_encoded', 'place_name_encoded', 'rooms', 'floor'
                                ,'distancia_obelisco','distancia_minima_subte','distancia_minima_privado',
                                'distancia_minima_publico']]
```

```python
test_to_predict = test_set.loc[:, ['property_type_encoded', 'state_name_encoded', 'place_name_encoded', 'rooms', 'floor'
                                ,'distancia_obelisco','distancia_minima_subte','distancia_minima_privado',
                                'distancia_minima_publico']]
```

```python
#De la documentacion de scikit
#init: determina como se inicializan los centroides iniciales(random/kmeans++ son dos ejemplos)
#n_clusters: la cantidad de clusters que se desean formar (Por default 8)
#n_init: el algoritmo se correra tantas veces como indique este parametro, y se quedara con el que arroje el mejor
#resultado
#max_iter: numero maximo de iteraciones que realizara el algoritmo para una sola corrida. (Por default 300)
kmeans = KMeans(init='k-means++', n_clusters= 100, n_init=1)
```

```python
#Los precios reales de las propiedades que uso para entrenar el algoritmo
train_prices = train.loc[:, 'price']
```

```python
predictions_sub_train = kmeans.fit(train_no_prices, train_prices).predict(sub_train_to_predict) 
predictions = kmeans.fit(train_no_prices, train_prices).predict(test_to_predict)
predictions = predictions.flatten()
```

```python
#Result almacena los clusters para los cual predijo el algoritmo que corresponde cada propiedad del set de test
result_sub_train = pd.DataFrame()
result_sub_train['cluster'] = predictions_sub_train
result = pd.DataFrame()
result['cluster'] = predictions
```

```python
real = train_sub_test.copy()
real['cluster'] = result_sub_train['cluster'].values.copy()
real[real['cluster'] == 8]
```

```python
result_sub_train['price_usd'] = real.groupby(['cluster'])['price'].transform('mean').values
real['price_usd'] = result_sub_train['price_usd'].values
price_by_cluster = real[['cluster','price_usd']]
```

```python
price_by_cluster.drop_duplicates(inplace=True)
price_by_cluster.sort_values(by='cluster',ascending=False)
price_by_cluster.head(10)
```

```python
final_prediction = real[['price','price_usd']]
score = mean_squared_error(real['price'].values, real['price_usd'].values)
score
```

```python
final_df = pd.merge(result,price_by_cluster,on='cluster',how = 'left')
#final_df.dropna(inplace = True)
final_df.head(10)
```
